```python
import json
f = open('./US-Stock-Symbols/nyse/nyse_full_tickers.json')
data = json.load(f)
#print(data)
symbols = []
# Save symbols into list
for i in data:
    symbols.append(i['symbol'])
```

```python
# Create yahoo compatible list of ticker symbols
symbols_yahoo = symbols

for i in range(len(symbols_yahoo)):
    symbol = symbols_yahoo[i]
    if "^" in symbol:
        symbols_yahoo[i] = symbols_yahoo[i].replace('^', '-P')

# print(symbols_yahoo)
        
```

```python
# Create reuters compatible list of tickers
symbols_reuters = symbols

for i in range(len(symbols_reuters)):
    symbol = symbols_reuters[i]
    if "^" in symbol:
        symbols_reuters[i] = symbols_reuters[i].replace('^', '_p')
        symbols_reuters[i] = symbols_reuters[i] + ".N"

print(symbols_reuters)
```

# Note

* On reuters tickers with ^ in the name have _p in place of ^
    * On nyse .N is added to the end
* On yahoo, ^ is replaced by a hyphen and P: -P

```python
from yahoofinancials import YahooFinancials

#first_10 = symbols_yahoo[:10]

tickers = symbols_yahoo

yahoo_financials = YahooFinancials(tickers)

balance_sheet_data_qt = yahoo_financials.get_financial_stmts('quarterly', 'balance')
income_statement_data_qt = yahoo_financials.get_financial_stmts('quarterly', 'income')


```

```python

# Calculates asset growth in a very, very ugly way

d = {}
for i in balance_sheet_data_qt.values():
    for j in i.keys():
        d[j] = {}
        # d[j]['retained_earnings'] = 
        print(len(i[j]))
        print("")

        if len(i[j]) > 0:
            # if i[j] is not None:
            keys = [key_val for key_val in i[j][0].keys()]
            key_newer = keys[0]

            keys2 = [key_val for key_val in i[j][1].keys()]
            key_later = keys2[0]

            print("keys: ", keys)

            values_newer = i[j][0][key_newer]
            values_later = i[j][1][key_later]
            print("Values: ", values_newer)
            print("Values  later = ", values_later)

            # print("Key: ", key)
            if 'totalAssets' in values_newer.keys():
                print(i[j][0])
                d[j]['asset_growth'] = values_newer['totalAssets'] - values_later['totalAssets']
                d[j]['total_stockholder_equity'] = values_newer['totalStockholderEquity']
                d[j]['retained_earnings'] = values_newer['retainedEarnings']
                d[j]['most_recent_data'] = key_newer
                d[j]['stock_exchange'] = "NYSE"
        
```

```python
# Values from income statement
for i in income_statement_data_qt.values():
    for j in i.keys():
        # d[j]['retained_earnings'] = 
        print(len(i[j]))
        print("")

        if len(i[j]) > 0:
            # if i[j] is not None:
            keys = [key_val for key_val in i[j][0].keys()]
            key_newer = keys[0]

            # keys2 = [key_val for key_val in i[j][1].keys()]
            # key_later = keys2[0]

            print("keys: ", keys)

            values_newer = i[j][0][key_newer]
            #values_later = i[j][1][key_later]
            print("Values: ", values_newer)
            # print("Values  later = ", values_later)

            # print("Key: ", key)
            if 'totalRevenue' in values_newer.keys():
                print(i[j][0])
                d[j]['previous_revenue'] = values_newer['totalRevenue']
```

```python
# Get market caps
market_caps = yahoo_financials.get_market_cap()
print(market_caps)
```

```python
# add to d dictionary
for key, value in market_caps.items():
    d[key]['market_cap'] = value
```
