```python
# Only works with output.json
def read_common_stock_symbols_from_file(file_path):
    """ Returns only tickers that are common stock and
    listed on NYSE """
    import json
    f = open(file_path)
    data = json.load(f)
    return [i['normalized_ticker'] for i in data if i['instrument_type'] == "COMMON_STOCK"]
```

```python
def get_stock_returns_previous_month(ticker):
    """ Returns stock's returns for the previous month 
    
    Remember that if the original ticker symbol contains a '^'
    symbol, you have to replace that with '-p' on Yahoo
    finance """
    ticker_symbol = validate_ticker(ticker)

    print("TICKER: ", ticker_symbol)

    from pandas_datareader import data as pdr

    import yfinance as yf
    yf.pdr_override() # <== that's all it takes :-)

    # Current date
    from datetime import datetime

    now = datetime.now() # current date and time    

    now_formatted = now.strftime('%Y-%m-%d')
    print("Now formatted: ", now_formatted)

    # download dataframe
    data = pdr.get_data_yahoo(ticker_symbol, start="2017-01-01", end=now_formatted)

    monthly_returns = data['Adj Close'].resample('M').ffill().pct_change()

    # Return second last element from list
    return monthly_returns[-2]
```

```python
def validate_ticker(ticker):
    """ Checks that the ticker is in correct format """
    # Seems to work
    if "^" in ticker:
        print("TICKER: ", ticker)
        ticker = ticker.replace('^', '-P')
    return ticker
```
